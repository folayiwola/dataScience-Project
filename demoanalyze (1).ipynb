```python
bsc_grads = adult[adult['education'] == 'Bachelors']
salary_Bsc = bsc_grads[bsc_grads['salary'] == '>50K']
len(salary_Bsc)
```

```python
msc_grads = adult[adult['education'] == 'Masters']
salary_Msc = msc_grads[msc_grads['salary'] == '>50K']
len(salary_Msc)
```

```python
phd_grads = adult[adult['education'] == 'Doctorate']
salary_Phd = phd_grads[phd_grads['salary'] == '>50K']
len(salary_Phd)
```

```python
sumabove_50k = len(salary_Bsc)+len(salary_Msc)+len(salary_Phd)
total_num = len(bsc_grads)+len(msc_grads)+len(phd_grads)
pabove_50k = sumabove_50k/total_num * 100
print(pabove_50k)
```
