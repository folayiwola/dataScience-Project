```python
#calculating the distribution of values in the date_crawled, ad_created, and last_seen columns as percentages
dates = autos['dateCrawled'].value_counts(normalize=True,dropna=False)
ads = autos['ad_created'].value_counts(normalize=True,dropna=False)
last = autos['lastSeen'].value_counts(normalize=True,dropna=False)
percent_dates = dates/len(dates)
percent_ads = dates/len(ads)
percent_last = dates/len(last)
percent_dates.sort_index()
percent_ads.sort_index()
percent_last.sort_index()
print(percent_dates)
print(percent_ads)
print(percent_last)
```

```python
#calculate the distribution of the remaining values after removing upper and lower bound values
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).head(6)
```

```python
autos['brand'].unique
autos['brand'].value_counts(normalize=True)
```

```python
#To aggregate across brands to understand mean price
brand_counts =autos['brand'].value_counts(normalize=True)
pop_brands = brand_counts[brand_counts > .05].index
print(pop_brands)
```

```python
brand_mean_prices = {}
brand_mean_mileage = {}

for brand in pop_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)
    

brand_mean_prices
```

```python
brand_mean_mileage = {}

for brands in pop_brands:
    brand_only = autos[autos['brand'] == brands]
    mean_mileage = brand_only['odometer_km'].mean()
    brand_mean_mileage[brands] = int(mean_mileage)
    
brand_mean_mileage
```
